$표제어\ 추출^{lemmatization}$ 또는 $어간\ 추출^{stemming}$을 수행하여 접사등을 제거한 이후에 텍스트 분류를 적용해야 하는지에 관한 것입니다. 

예를 들어 "나는 학교에 가요"라는 원문이 있을때, 여기서 표제어 또는 어간 추출을 하면 다음과 같은 형태의 문장이 됩니다.

<br></br>

|단계|문장|
|---|---|
|원문|나는 학교에 가요.|
|전처리|나 는 학교 에 가 요 .|
|추출|나 학교 가 .|

<br></br>

원문에 비교하면 희소성 문제를 훨씬 더 줄일 수 있을 것입니다. 표제어 또는 어간 추출 모듈의 정책에 따라 다르겠지만 "나 학교 가 ."와 같은 추출 결과를 얻을 수 있는 원문의 종류는 굉장히 많을 것입니다

<br></br>

|번호|문장|
|---|---|
|1|나는 학교에 가요.|
|2|나도 학교에 가요.|
|3|나는 학교를 가요.|
|4|나만 학교에 가요.|
|5|나도 학교를 가요.|

<br></br>

코퍼스가 부족한 상황에서는 이처럼 어간이나 표제어가 같은 문장에 대해 같은 샘플로 취급하여 희소성 문제에서 어느 정도 타협을 볼 수 있습니다.

하지만 표제어 추출 및 어간 추출이 반드시 정석이라고 볼 수는 없습니다. 예를 들어 다음과 같은 문장에서 감성 분석을 수행한다면 표제어 추출 또는 어간 추출을 할 경우에는 텍스트 분류가 어려울 것입니다.

<br></br>

|원문|추출 후|정답|
|---|------|---|
|나는 학교에 가요.|나 학교 가 .| 긍정 또는 중립|
|나만 학교에 가요.|나 학교 가 .| 부정|
